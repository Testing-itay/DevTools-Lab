# AI Model Experiments

```python
import openai
client = openai.OpenAI()
response = client.chat.completions.create(model="gpt-4", messages=[{"role": "user", "content": "Hello"}])
print(response.choices[0].message.content)
```

```python
import anthropic
client = anthropic.Anthropic()
response = client.messages.create(model="claude-3-sonnet-20240229", max_tokens=1024, messages=[{"role": "user", "content": "Hello"}])
print(response.content[0].text)
```
